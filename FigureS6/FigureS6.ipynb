```python
import matplotlib.pyplot as plt
import numpy as np

def reader(fh):
    data = []
    for line in fh:
        if line[0] == "#":
            print(line, end="")
            continue
        data.append([float(x) for x in line.split()])
    return np.array(data)

fig = plt.figure()
ax = plt.gca()

ax.set_xlim(3,100)
ax.set_xlabel(r"$r$ / 0.1 nm",fontsize=14)
ax.set_ylim(0.2,10)
ax.set_ylabel(r"$\sigma(r)$" + r" / kJ mol$^{-1}$",fontsize=18)

ice1c64 = reader(open("1css.d"))
plt.loglog(ice1c64[:,0], ice1c64[:,2], label="64")
plt.axvline(25.52729/4)

ice1c512 = reader(open("1cs.d"))
plt.loglog(ice1c512[:,0], ice1c512[:,2], label="512")
plt.axvline(25.52729/2)

ice1c4096 = reader(open("1cmm.d"))
plt.loglog(ice1c4096[:,0], ice1c4096[:,2], label="4096")
plt.axvline(51.05457/2)

ice1cxxx = reader(open("sd/1c.d"))
plt.loglog(ice1cxxx[:,0], ice1cxxx[:,2], label="262144")
plt.axvline(204.21828/2)
plt.legend()

fig.savefig("FigureS7.pdf", bbox_inches="tight")
```
